# üìä 1. Pr√©sentation de la biblioth√®que Plotly

## 1.1 Objectifs de Plotly

Plotly est une **biblioth√®que de visualisation de donn√©es interactive** pour Python. Son objectif principal est de permettre aux data scientists, analystes et ing√©nieurs d‚Äôexplorer et de pr√©senter leurs donn√©es sous forme de graphiques dynamiques, professionnels et facilement interpr√©tables.

---

## 1.2 Architecture g√©n√©rale de Plotly

Plotly est organis√© en plusieurs composants principaux :

1. **plotly.graph_objects**
   - Composants graphiques bas niveau permettant un contr√¥le total sur les √©l√©ments des graphes.

üîß **Fonctionnement interne (simplifi√©) :**

- Plotly g√©n√®re des objets graphiques Python.
- Ceux-ci sont convertis en **code HTML + Javascript (Plotly.js)**.
- Ce code s‚Äôaffiche dans un notebook Jupyter ou un navigateur, rendant le graphique interactif.

## 2.1 Modules principaux

### ‚úÖ plotly.express (px)

üîß **Description :**

- `plotly.express` est le **module haut niveau** de Plotly.
- Il permet de cr√©er rapidement des graphiques courants (scatter plot, bar chart, line plot, box plot, choropleth map, etc.) en **une seule ligne de code**.

üîë **Utilisation principale :**

- Id√©al pour l‚Äô**exploration rapide des donn√©es**.  
- G√©n√®re automatiquement la figure et affiche un graphique interactif.

üíª **Exemple pratique :**



````python
import plotly.express as px

# Charger un dataset exemple
df = px.data.gapminder()

# Filtrer l'ann√©e 2007
df_2007 = df[df["year"] == 2007]

# Cr√©er un scatter plot interactif
fig = px.scatter(
    data_frame=df_2007,
    x="gdpPercap",
    y="lifeExp",
    color="continent",
    size="pop",
    hover_name="country",
    log_x=True,
    size_max=60,
    title="GDP per Capita vs Life Expectancy (2007)"
)

fig.show()

````

 ## plotly.graph_objects (go)
üîß Description :

plotly.graph_objects est le module bas niveau de Plotly.

Il donne un contr√¥le total sur la construction du graphique, chaque √©l√©ment peut √™tre modifi√© individuellement.

üîë Utilisation principale :

Cr√©er des visualisations hautement personnalis√©es.

Ajouter plusieurs traces (courbes, barres, surfaces) et modifier le layout en d√©tail.

üíª Exemple pratique :

````python
import plotly.graph_objects as go

# Cr√©er une figure vide
fig = go.Figure()

# Ajouter une trace scatter
fig.add_trace(go.Scatter(
    x=df_2007["gdpPercap"],
    y=df_2007["lifeExp"],
    mode="markers",
    marker=dict(size=10, color="red"),
    text=df_2007["country"]
))

# Personnaliser le layout
fig.update_layout(
    title="GDP per Capita vs Life Expectancy (2007) - graph_objects",
    xaxis_title="GDP per Capita",
    yaxis_title="Life Expectancy",
    template="plotly_white"
)

fig.show()
````

üéØ R√¥le : construire des graphiques personnalis√©s et multi-traces adapt√©s aux pr√©sentations professionnelles et dashboards interactifs.

# üí∞ Partie 3 : Mise en ≈ìuvre pratique ‚Äì Analyse comparative de 10 indices boursiers mondiaux

## üéØ Objectif

Comparer l‚Äô√©volution de **10 indices boursiers mondiaux** (Dow Jones, Nasdaq, CAC 40, DAX, FTSE 100, Nikkei 225, Shanghai Composite, Sensex, S&P 500, Euro Stoxx 50) de 2020 √† 2024 gr√¢ce √† **Plotly**.

---

## üîß 3.1 Installation et importation des biblioth√®ques



````python
# Installer yfinance si n√©cessaire
!pip install yfinance


````

````python
# Importer les biblioth√®ques
import yfinance as yf
import pandas as pd
import plotly.express as px
````

## 3.2 T√©l√©chargement des donn√©es

````python
# Liste des indices et leurs tickers Yahoo Finance
indices = {
    "^DJI": "Dow Jones",
    "^IXIC": "Nasdaq",
    "^GSPC": "S&P 500",
    "^FTSE": "FTSE 100",
    "^GDAXI": "DAX",
    "^FCHI": "CAC 40",
    "^N225": "Nikkei 225",
    "000001.SS": "Shanghai Composite",
    "^BSESN": "Sensex",
    "^STOXX50E": "Euro Stoxx 50"
}

# T√©l√©charger les donn√©es de 2020 √† 2024
data = yf.download(list(indices.keys()), start="2020-01-01", end="2024-12-31")["Close"]

# Renommer les colonnes avec les noms clairs
data.rename(columns=indices, inplace=True)

# Aper√ßu des donn√©es
data.head()
````

## 3.3 Pr√©paration des donn√©es pour Plotly

````python
# R√©initialiser l'index
df = data.reset_index()

# Transformer en format long
df_melted = df.melt(id_vars=["Date"], var_name="Indice", value_name="Close")

# Aper√ßu des donn√©es transform√©es
df_melted.head()
````

## 3.4 Visualisation interactive comparative

````python
# Cr√©er un graphique line plot interactif
fig = px.line(
    df_melted,
    x="Date",
    y="Close",
    color="Indice",
    title="üìà √âvolution des principaux indices boursiers (2020-2024)",
    labels={"Close":"Cours de cl√¥ture ($)", "Date":"Date"}
)

fig.show()
````

````python
# Calcul du rendement quotidien et annualis√©
returns = data.pct_change()
annual_return = returns.mean() * 252
volatility = returns.std() * (252 ** 0.5)

# Cr√©er un DataFrame r√©capitulatif
risk_return = pd.DataFrame({
    "Annual Return (%)": annual_return * 100,
    "Volatility (%)": volatility * 100
})

risk_return.reset_index(inplace=True)
risk_return.rename(columns={"index":"Indice"}, inplace=True)

# Visualisation Risk vs Return
fig = px.scatter(
    risk_return,
    x="Volatility (%)",
    y="Annual Return (%)",
    text="Ticker",
    title="‚öñÔ∏è Risk vs Return des indices boursiers (2020-2024)"
)
fig.show()
````
